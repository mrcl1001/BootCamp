## 2. Singular Value Decomposition

Este ejercicio pondr치 a prueba tu habilidad para usar Singular Value Decomposition para comprimir una imagen.

**Problema:** Usar `SVD` para comprimir una imagen en blanco y negro.

La imagen que deberas usar es la siguiente:

```python
import matplotlib.pyplot as plt
from scipy import misc
%matplotlib inline

# Load image
A = misc.face(gray=True)

plt.imshow(A, cmap=plt.cm.gray)
```

Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
import numpy as np

def sse_score(X, X_hat):
    return np.sum((X - X_hat)**2)

X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat)
print(round(sse, 5))
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVM`.

```python
def svm(X):
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    return U, np.diag(s), Vt 

X = np.array([[1, 2], [3, 4]])
U2, S2, Vt2 = svm(X) 


print("U = array(", np.round(U2, 2), ")")
print("\nS = array(", np.round(S2, 2), ")")
print("\nVt = array(", np.round(Vt2, 2), ")")
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
def reconstruction(U, S, Vt):
    for values in range(1, S.shape[0]):
        U_temp  =  U[0:U.shape[0], 0:values]
        S_temp  =  S[0:values,     0:values]
        Vt_temp = Vt[0:values,     0:Vt.shape[1]]

    return U_temp.dot(S_temp.dot(Vt_temp))


U  = np.array([[-0.40455358, -0.9145143 ], [-0.9145143, 0.40455358]])
S  = np.array([[5.4649857 , 0.], [0., 0.36596619]])
Vt = np.array([[-0.57604844, -0.81741556], [ 0.81741556, -0.57604844]])
X_hat = reconstruction(U, S, Vt)
print(X_hat)
```

Calcula el error de reconstrucci칩n usando la funci칩n `sse` que has programado anteriormente.

```python
sse = sse_score(X, X_hat)
print(f"El error de reconstrucci칩n es: { round(sse, 4) }")
```

Una vez que hemos programado todas las funciones necesarias para realizar `SVM` y medir el error de reconstrucci칩n, podemos proceder a realizar la compresi칩n de la imagen. Esta [p치gina web](http://timbaumann.info/svd-image-compression-demo/) te ayudar치 a repasar y a entender como calcular la compresi칩n.

Debes usar la siguiente imagen: 

```python
# Load image
A = misc.face(gray=True)

plt.imshow(A, cmap=plt.cm.gray)
```

```python
def image_compression(A, n_components):
    U_comp, S_comp, Vt_comp = svm(A)
    A_hat_comp = reconstruction(U_comp, S_comp, Vt_comp)    
    sse_comp   = sse_score(A, A_hat_comp)    
    ratio_comp = (A.shape[1]*n_components + n_components + A.shape[0]*n_components)/(A.shape[1] * A.shape[0])

    return A_hat_comp, sse_comp, ratio_comp

A_hat, sse, comp_ratio = image_compression(A, n_components=50)
print("El error de reconstrucci칩n es: ", sse)
print("El ratio de compresi칩n es: ", comp_ratio)
print("La Im치gen reconstruida es : \n", plt.imshow(A_hat, cmap=plt.cm.gray))
```

El siguiente paso es construir una funci칩n que recibe un `data_frame`, el nombre de una columna, un valor m칤nimo y un valor m치ximo `cutoffs`. Nos devuelve un dataframe excluyendo las filas donde el valor de la columna indica est치 fuera de los valores m칤nimos y m치ximos que le hemos indicado.

```python
def column_cutoff(data_frame, cutoffs):
    columns_df  = []
    
    for i in cutoffs:
        columns_df.append(i[0])
    
    data_frame = data_frame[columns_df]

    for i in cutoffs:
        data_frame = data_frame[data_frame[i[0]] > i[1]]
        data_frame = data_frame[data_frame[i[0]] < i[2]]

    return data_frame

data_frame = read_to_df('./data/train.csv')
# Remove data points with SalePrice < $50,000
# Remove data points with GrLiveAre > 4,000 square feet
cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
selected_data = column_cutoff(data_frame, cutoffs)

print(selected_data)
```

#### M칤nimos Cuadrados / Least Squares

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
import numpy as np

def least_squares_weights(input_x, target_y):
    #Paso 1
    n1 = input_x.shape[0]
    d1 = input_x.shape[1]
    if n1 < d1:
        input_x = np.transpose(input_x)
        n1 = input_x.shape[0]

    n2 = target_y.shape[0]
    d2 = target_y.shape[1]
    if n2 < d2:
        target_y = np.transpose(target_y)

    #Paso 2
    new_column = np.ones((1, n1))
    X = np.insert(input_x, input_x.shape[1], np.array(new_column), 1)

    #Paso 3
    Xt       = np.transpose(X)
    XtX      = np.dot(Xt, X)
    XtX_1    = np.linalg.inv(XtX)
    XtX_1Xt  = np.dot(XtX_1, Xt)
    XtX_1Xty  = np.dot(XtX_1Xt, target_y)

    return XtX_1Xty

training_x = np.array([[1710, 1262, 1786, 
                        1717, 2198, 1362, 
                        1694, 2090, 1774, 
                        1077], 
                        [2003, 1976, 2001, 
                        1915, 2000, 1993, 
                        2004, 1973, 1931, 
                        1939]])

training_y = np.array([[208500, 181500, 223500, 
                        140000, 250000, 143000, 
                        307000, 200000, 129900, 
                        118000]])

weights = least_squares_weights(training_x, training_y)

print(weights)                     
print(weights[1][0])
```

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

Una vez que tenemos una lista de conjuntos por cada par de p치ginas web, podemos convertir el texto de la p치gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci칩n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p치gina web 1
        tokens_web_2 -- Contenido tokenizado de p치gina web 2
        set_web1 -- Conjunto de palabras de la p치gina web 1
        set_web2 -- Conjunto de palabras de la p치gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci칩n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
```

```python
def websites_sim(url_ls):
    """Funci칩n para calcular la similaridad entre p치ginas web
    
    Argumentos:
        url_ls -- Listas de p치ginas web
        
    Ejemplo:
        url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
        similarity_ls = websites_sim(url_ls)  
    """
    
    url2content = get_content(url_ls)
    
    # Creamos un diccionario donde cada url tendr치 su contenido tokenizado y su conjunto
    url_dict = {}
    for url, content in url2content.items():
        toks, sets = tokenizer(content)
        url_dict[url] = {'tokens': toks,
                        'unique_tokens': sets}
    
    # Calculamos todas las combinaciones posibles de las direcciones de las p치ginas web
    comb_ls = combinations(url_ls)

    # Usando comb_ls y la funci칩n `set2vector()` convertimos cada p치gina web a vectores
    print("Similaridad: ")
    for el in comb_ls:
        # Obtenemos los tokens y el conjunto para cada p치gina web
        token_1 = url_dict[el[0]]['tokens']
        token_2 = url_dict[el[1]]['tokens']
        set_1 = url_dict[el[0]]['unique_tokens']
        set_2 = url_dict[el[1]]['unique_tokens']
        array_web1, array_web2 = set2vector(token_1, token_2, set_1, set_2)
        similarity = cosine_similarity(array_web1, array_web2)
        
        print("{} vs {} - {}".format(el[0], el[1], round(similarity, 3)))

                      
url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
'https://es.wikipedia.org/wiki/Canis_lupus',
'https://es.wikipedia.org/wiki/Felis_silvestris_catus']

similarity_ls = websites_sim(url_ls) 
```
